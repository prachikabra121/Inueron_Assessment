```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generating a sample dataset
np.random.seed(42)
data = {'Variable_1': np.random.rand(100) * 10,
        'Variable_2': np.random.rand(100) * 5 + 2}

df = pd.DataFrame(data)

# Calculating Pearson correlation coefficient
correlation_coefficient = df['Variable_1'].corr(df['Variable_2'])

# Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Variable_1', y='Variable_2', data=df)
plt.title(f'Scatter Plot with Correlation Coefficient: {correlation_coefficient:.2f}')
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.show()

print(f'Pearson Correlation Coefficient: {correlation_coefficient:.2f}')
```
